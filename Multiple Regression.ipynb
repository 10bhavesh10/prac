```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('mul_reg.csv')
```

```python
df.head()
```

```python
state=pd.get_dummies(df['State'])
df=df.join(state)
df=df.drop('State',axis=1)
```

```python
df.head()
```

```python
df.columns
```

```python
columns=['R&D Spend', 'Administration', 'Marketing Spend','California', 'Florida', 'New York', 'Profit']
df=df[columns]
```

```python
df.head()
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

For any ml model x->2-D  y->1-D

```python
x=pd.DataFrame(x)
x.columns=['R&D Spend', 'Administration', 'Marketing Spend',
       'California', 'Florida', 'New York']
x['intercept']=1
x.head()
```

```python
vif=pd.DataFrame()
vif['variable']=x.columns
```

```python
vif
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif['values']=[variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
```

```python
vif
```

```python
x=x.drop('California',axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
print(reg.coef_,reg.intercept_)
```

```python
y_pred=reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
import math
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(math.sqrt(mean_squared_error(y_test,y_pred)))
```
